```python
import pymupdf4llm
from datasets import load_dataset
import json
```

```python
lst20_dataset = load_dataset('lst-nectec/lst20', trust_remote_code = True, data_dir="/content/LST20_Corpus")
```

```python
lst20_dataset_ner_df = lst20_dataset['train'].to_pandas()[['tokens', 'ner_tags']]
lst20_dataset_ner_df_test = lst20_dataset['test'].to_pandas()[['tokens', 'ner_tags']]
lst20_dataset_ner_df
```

```python
lst20_ner_index = [
  "O",
  "B_BRN",
  "B_DES",
  "B_DTM",
  "B_LOC",
  "B_MEA",
  "B_NUM",
  "B_ORG",
  "B_PER",
  "B_TRM",
  "B_TTL",
  "I_BRN",
  "I_DES",
  "I_DTM",
  "I_LOC",
  "I_MEA",
  "I_NUM",
  "I_ORG",
  "I_PER",
  "I_TRM",
  "I_TTL",
  "E_BRN",
  "E_DES",
  "E_DTM",
  "E_LOC",
  "E_MEA",
  "E_NUM",
  "E_ORG",
  "E_PER",
  "E_TRM",
  "E_TTL"
]
```

```python
lst20_ner_guideline = pymupdf4llm.to_markdown('/content/superai_ss5_ner/2008.05055v1.pdf', pages=[16,17,18,19]) # Index start at 0, and NER guideline is at page 17-20
```

```python
print(lst20_ner_guideline)
```

```python
lst20_ner_index_prompt = "\n".join([f"index: {i}: {item}" for i, item in enumerate(lst20_ner_index)])
print(lst20_ner_index_prompt)
```

```python
import os
import google.generativeai as genai
from google.colab import userdata


genai.configure(api_key=userdata.get('GEMINI_API_KEY'))

# Create the model
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 40,
  "max_output_tokens": 8192,
  # "response_mime_type": "text/plain",
  "response_mime_type": "application/json",
}

model = genai.GenerativeModel(
  model_name="gemini-2.0-flash-exp",
  # model_name="gemini-2.0-flash-thinking-exp-1219",
  generation_config=generation_config,
)

```

```python
lst20_ner_examples_prompts = []
for i, row in lst20_dataset_ner_df.iloc[:20].iterrows():
  examples = []
  for token, ner in zip(row['tokens'], row['ner_tags']):
    item = {"token": token, "ner": ner}
    examples.append(repr(item))
  lst20_ner_examples_prompts.append(f"[{', '.join(examples)}]")

lst20_ner_examples_prompt = "\n".join(lst20_ner_examples_prompts)
print(lst20_ner_examples_prompt)
```

```python
def ask_ner(sentence: str, expected_len: int):
  chat_session = model.start_chat(
  history=[
      {
        "role": "user",
        "parts": [
          lst20_ner_guideline + \
          lst20_ner_index_prompt + \
          lst20_ner_examples_prompt + \
          "From this guideline please tag the following sentence and output as json list of dict that has key token: str and ner: integer \n" + \
          f"Your result len should be {expected_len} \n" + \
          "sentence:"
        ],
      },
  ]
)

  response = chat_session.send_message(sentence)
  result = json.loads(response.text)
  if len(result) != expected_len:
    fix_prompt = f"Please correct the mistake. Your result len should be {expected_len} not {len(result)} \n sentence:" + sentence
    result = chat_session.send_message(fix_prompt)
    result = json.loads(response.text)
    if len(result) > expected_len:
      print('Error: len(result) > expected_len')
      result = result[:expected_len]
    elif len(result) < expected_len:
      print('Error: len(result) < expected_len')
      result = result + [0] * (expected_len - len(result))

  result_processed = []
  for item in result:
    result_processed.append(item['ner'])
  return result_processed
```

```python
lst20_dataset_ner_df_test_sample = lst20_dataset_ner_df_test.iloc[:100]
```

```python
str(lst20_dataset_ner_df_test_sample['tokens'].iloc[30])
```

```python
ask_ner(str(lst20_dataset_ner_df_test_sample['tokens'].iloc[30]), len(lst20_dataset_ner_df_test_sample['tokens'].iloc[30]))
```

```python
from tqdm.auto import tqdm
import time

ner_results = []
for i, row in tqdm(lst20_dataset_ner_df_test_sample.iterrows(), total = len(lst20_dataset_ner_df_test_sample)):
  expected_len = len(row['tokens'])
  try:
    ner_results.append(ask_ner(str(row['tokens']), expected_len))
  except Exception:
    import traceback
    traceback.print_exc()

    time.sleep(30)
    ner_results.append(ask_ner(str(row['tokens']), expected_len))
  finally:
    time.sleep(6)
```

```python
import pandas as pd
result_df = pd.Series(ner_results)
```

```python
len(result_df.explode().tolist())
```
