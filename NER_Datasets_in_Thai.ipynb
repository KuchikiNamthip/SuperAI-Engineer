```python
from transformers import RobertaTokenizerFast, RobertaForTokenClassification
from attacut import tokenize as attacut_tokenize
import torch

tokenizer = RobertaTokenizerFast.from_pretrained("lst-nectec/HoogBERTa-NER-lst20")
hoog_model = RobertaForTokenClassification.from_pretrained("lst-nectec/HoogBERTa-NER-lst20")
```

```python
from transformers import pipeline

nlp = pipeline('token-classification', model=hoog_model, tokenizer=tokenizer, aggregation_strategy="none")
```

```python
# From https://github.com/ThilinaRajapakse/simpletransformers/blob/919f69d74b5a10ced24d8eecbe73d31409bff95d/simpletransformers/ner/ner_utils.py#L241
from typing import List

pad_token_label_id = 99

def map_custom_tokenized_sentence_to_ner_tokenizer(targeted_tokens: List[str], targeted_ners: List[int]):
  tokens = []
  label_ids = []

  for i, (word, label) in enumerate(zip(targeted_tokens, targeted_ners)):
      word_tokens = tokenizer.tokenize(word)
      if (
          word_tokens
      ):  # avoid non printable character like '\u200e' which are tokenized as a void token ''
          tokens.extend(word_tokens)
      else:
          word_tokens = tokenizer.tokenize(tokenizer.unk_token)
          tokens.extend(word_tokens)
      label_ids.extend([label] + [pad_token_label_id] * (len(word_tokens) - 1))

  tokens_full = [tokenizer.cls_token] + tokens + [tokenizer.sep_token]
  label_ids_full = [-100] + label_ids + [-100]

  return label_ids_full, tokens_full
```

```python
label_ids_full, tokens_full = map_custom_tokenized_sentence_to_ner_tokenizer(targeted_tokens, targeted_ners)
label_ids_full, tokens_full
```
